### Webscraping

```python
from bs4 import BeautifulSoup #This module allows you scrape the web

import urllib.request as ur #sends a request to a website

url = "https://www.theguardian.com/international" #save the url in a variable

#send request to the webpage
Query = ur.urlopen(url)

#Parse
mySoup = BeautifulSoup(Query)

mySoup
```

```python
mytag = mySoup.find("h") #find a tag

print(mytag)

mytag.text.strip() #strip out all tags from text
```

```python
mytag2 = mySoup.findAll("h3") #extract all tags

mytag2
```

```python
mynum = [1, 3, 4, 3]

for i in mynum:
    if i == 3:
        mynum.remove(i)
```

```python
mynum
```

```python
mytag2.text.strip() #text strip works for only one element at a time
```

```python
content = [] #create an empty list to save the result of the loop

for i in mytag2:
    content.append(i.text.strip())
content
#content
```

```python
filename = "bbcfrontnews.txt" #write

with open(filename, "w") as bbc:
    for i in content:
        bbc.write(i + "\n")
```

```python
#Assignment: Write a Python function, that returns the mean,min, max of revenue from the dataframe df

df = pd.read_excel("SalesForecast_RawData.xlsx")

df.head()
```

```python
def minmax(): #global variable approach
    a = df[["Revenue"]].min()
    b = df[["Revenue"]].max()
    c = df[["Revenue"]].mean()
    return a, b, c

minmax()
```

```python
def minmax(dataframe): #input approach
    a = dataframe[["Revenue"]].min()
    b = dataframe[["Revenue"]].max()
    c = dataframe[["Revenue"]].mean()
    return a, b, c

minmax(dataframe = df)
```
